```python
#import libraries
import torch
import torch.nn as nn
```

```python
# create a  model and weight matrices sizes
# build a model
aModel = nn.Sequential(
    nn.Linear(10,14),
    nn.Linear(14,19),
    nn.Linear(19,8),
        )
aModel
```

```python
# print the weight matrices sizes of each layer
for i in range(len(aModel)):
    print(aModel[i].weight.shape)
```

# Build a Model with inconsistent layer shapes

```python
M2 = nn.Sequential(
    nn.Linear(10,14),
    nn.Linear(14,9),
    nn.Linear(19,8),
        )
for i in range(len(M2)):
    print(M2[i].weight.shape)
```

```python
# test both models with fake data

# generate the data
nsamples = 5
nfeatures = 10

fakedata = torch.randn(nsamples,nfeatures)
  
```

```python
# test the first model
aModel(fakedata).shape
```

```python
# test the second Model
M2(fakedata).shape # it will not make the sense because the weights are not correct.
```
