```python
# import libraries
import numpy as np
import torch
```

# Using Numpy

```python
# create a vector
v = np.array([1,40,2,-3])

# find the minimum and maximum values
minval = np.min(v)
maxval = np.max(v)

print("Min,max: %g,%g" %(minval,maxval))


# now for the argmin/max
minidx = np.argmin(v)
maxidx = np.argmax(v)

print('Min,max indices: %g,%g' %(minidx,maxidx)), print(' ')

#confirm
print(f'Min val is { v[minidx] }, max val is { v[maxidx] }')
```

```python
# repeat with matrix
M = np.array([[0,1,10], [20,8,5]])
print(M), print(' ')

#various minima in this matrix
minvals1 = np.min(M) # Minimum for entire matrix
minvals2 = np.min(M, axis=0)
minvals3 = np.min(M,axis=1)

# print
print(minvals1)
print(minvals2)
print(minvals3)
```

```python
# various minima in this matrix
minidx1 = np.argmin(M)
minidx2 = np.argmin(M,axis=0)
minidx3 = np.argmin(M,axis=1)

#print 
print(M),print(' ')
print(minidx1)
print(minidx2)
print(minidx3)
```

# Using pytorch

```python
# create a  vector
v = torch.tensor([1,40,2,-3])

# find the maximum and minimum values
minval = torch.min(v)
maxval = torch.max(v)

print("Min ,max: %g,%g" %(minval,maxval))

#now for argmin/max

minidx = torch.argmin(v)
maxidx = torch.argmax(v)

print("Min,max indices:%g,%g" %(minidx,maxidx) ), print(' ')

#confirm
print(f'Min val is {v[minidx]},max val is {v[maxidx]}')

# confirm
print(f'Min val is { v[minidx] }, max val is { v[maxidx] }')
```

```python
# repeat with matrix
M = torch.tensor([[0,1,10],[20,8,5]])
print(M),print(' ')

# various minima in this matrix
min1 = torch.min(M) # Minimum for entire matrix
min2 = torch.min(M,axis=0)
min3 = torch.min(M,axis=1)

# print
print(min1), print(' ')
print(min2), print(' ')
print(min2.values)
print(min2.indices)
```
